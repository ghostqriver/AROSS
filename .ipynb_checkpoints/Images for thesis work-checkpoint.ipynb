```python
def gd(x, mu=0, sigma=0.1):
    '''
    sigma: std
    '''
    left = 1 / (np.sqrt(2 * math.pi) * np.sqrt(sigma))
    right = np.exp(-(x - mu)**2 / (2 * sigma))
    return left * right

x = np.arange(-1.5, 1.5, 0.001) 
```

```python
y = gd(x, sigma=0.2)
```

```python
keep_oversamplers = ['ORIG','S','AS','DBS','CS','KS','SOMO','AROS','AROSS']
alphas = {1:'a',2:'b',3:'c',4:'d'}
marks = {'ORIG':'.','S':'^','AS':'+','DBS':'2','CS':'x','KS':'$\heartsuit$','SOMO':'$*$','AROS':'o','AROSS':'s'}
colors = {'ORIG':'r','S':'#FF7F0E','AS':'#17BECF','DBS':'#7F7F7F','CS':'#9467BD','KS':'#8C564B','SOMO':'#E377C2','AROS':'g','AROSS':'b'}
```

```python
import warnings
import pandas as pd
import os 
warnings.filterwarnings("ignore") 

file = 'images/paper/svm_skewness.xlsx'
results = pd.read_excel(file,sheet_name=None,index_col=0)

plt.figure(figsize=(16,6),dpi = 200)
for subfig,sheet_name in enumerate(results):
    resultsheet = results[sheet_name][keep_oversamplers]
    oversamplers = resultsheet.columns
    x_axis = resultsheet.index
    
    plt.subplot(1,4,subfig+1)

    for oversampler in oversamplers:
        result = resultsheet[oversampler].values
        if oversampler in ['AROS','AROSS','S']:
            facecolors='none'
            edgecolors=colors[oversampler]
        else:
            facecolors = colors[oversampler]
            edgecolors = colors[oversampler]
        if oversampler == 'CS':
            s = 80
        else:
            s = 100
        plt.scatter(x_axis,result,label=oversampler,marker=marks[oversampler],s=s,facecolors=facecolors,edgecolors=edgecolors)
        plt.title('('+alphas[subfig+1]+') '+sheet_name+'   ',y=-0.16,font={'family':'Times New Roman', 'size':16})
        plt.xticks(font={'family':'Times New Roman', 'size':14})
        plt.yticks(font={'family':'Times New Roman', 'size':12})

plt.subplot(1,4,1)
plt.legend(loc='lower left',prop={'family':'Times New Roman', 'size':12})
plt.savefig(file.split('.')[0]+'.png',bbox_inches='tight', pad_inches = 0)
```

Oversampler baseline

```python
def show_baseline(dataset,random_state=None,pos_label=None,img_name=None,**args):
    
    path = 'test/'
    make_dir(path)

    if img_name == None:
        fn = ''
        for str_ in [j[0]+str(j[1]) for j in [(i,args['args'][i]) for i in args['args']]]:
            fn = fn  + str_ +'_'
        # version = time.strftime('%m%d_%H%M%S')
        img_name  = fn+dataset+'_k='+str(random_state)+'.png'

    num_models = len(models)
    num_columns = 4
    num_rows = math.ceil(num_models/num_columns)
    
    plt.figure(figsize=(5*num_columns,5*num_rows))
    
    
    for ind,model in enumerate(models):

        X,y = read_data(dataset)

        X_train,X_test,y_train,y_test = split_data(X,y)

        pos_label = get_labels(y_test)[0]
        if model == 'cos':
            X_oversampled,y_oversampled,_,_ = oversampling(model,X_train,y_train,args['args']) 
        else:
            X_oversampled,y_oversampled = oversampling(model,X_train,y_train,args['args'])

        plt.subplot(num_rows,num_columns,ind+1)
        
        V.show_oversampling(X_train,y_train,X_oversampled,y_oversampled)
        plt.title(model)
        
    plt.savefig(path+img_name)
    print("The image stored in",path+img_name)
    plt.show()
```

```python
model_name = {'original':'ORIG','smote':'SMO','adasyn':'ADAS','db_smote':'DB_SMO','cure_smote':'CU_SMO','kmeans_smote':'KM_SMO','somo':'SOMO','AROS':'AROS','AROSS':'AROSS'}
```

```python
from COS_Funcs.utils import *
import COS_Funcs.utils.visualize as V
from COS_Funcs.baseline.baseline import *
from COS_Funcs.cos import cos
from COS_Funcs.cos import optimize
import warnings
warnings.filterwarnings("ignore") 

# dataset = 'Datasets/sampledata_new_2.csv'
dataset = 'toy4.csv'
models = ['original','smote','adasyn','db_smote','cure_smote','kmeans_smote','somo','AROS','AROSS']

num_models = len(models)
num_columns = 3
num_rows = math.ceil(num_models/num_columns)
plt.figure(figsize=(5*num_columns,5*num_rows),dpi=200)

X,y = read_data(dataset)
# X_train,X_test,y_train,y_test = split_data(X,y,random_state=22)
# X_train,X_test,y_train,y_test = read_fold(dataset,2) # 1 2
X_train,X_test,y_train,y_test = X,X,y,y

pos_label = get_labels(y_test)[0]

linkage = 'average'
# linkage = linkages[base_file(dataset)]
N = 80
# N = 25

for ind,model in enumerate(models):
    if model == 'AROS':
        X_oversampled,y_oversampled = cos.COS(X_train,y_train,N,linkage=linkage,all_safe_weight=1)[:2]
    elif model == 'AROSS':
        _,_,_,_,_,X_oversampled,y_oversampled = optimize.choose_alpha(X_train,y_train,X_test,y_test,'knn',['recall'],N,linkage,all_safe_weight=1)
    else:
        X_oversampled,y_oversampled = do_oversampling(model,X_train,y_train) 
    plt.subplot(num_rows,num_columns,ind+1)
    plt.xticks([],alpha=0)
    plt.yticks([],alpha=0)
    V.show_oversampling(X_train,y_train,X_oversampled,y_oversampled)
    plt.title(model_name[model],y=-0.16,font={'family':'Times New Roman', 'size':16})
plt.subplot(num_rows,num_columns,1)
# plt.legend(loc='lower left',prop={'family':'Times New Roman', 'size':14})

plt.savefig('images/'+os.path.basename(dataset).split('.')[0]+'.png',bbox_inches='tight')
```
