Multiple Scales in Gaussian

```python
import COS_Funcs.generate as G
import COS_Funcs.cos as cos
import COS_Funcs.baseline as baselines
import COS_Funcs.cure as cure
import COS_Funcs.visualize as V

import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = baselines.read_data('Dataset/','Sampledata_new_1')
```

```python
clusters,all_reps,num_reps = cure.Cure(X,30,10,0.2)
areas,min_all_safe_area,min_half_safe_area = cos.safe_areas(X,all_reps,y) 
```

```python
plt.figure(figsize=(5,5))
area = min_half_safe_area[0]
V.show_single_area(area)
plt.legend(loc='lower right')
plt.show()
```

```python
area = min_all_safe_area[0]
plt.figure(figsize=(15,10))
for i,scale in enumerate([0.1,0.3,0.9,2.7,10,100]):

    new_points = G.Gaussian_Generator(area,20,scale)
    plt.subplot(2,3,i+1)
    V.show_single_area(area,new_points)
    plt.title(str(scale))
    plt.legend(loc='lower left')
    
plt.show()
```

```python
new_points = G.Gaussian_Generator(area,50)
plt.figure(figsize=(10,10))
V.show_single_area(area,new_points)
plt.legend(loc='lower left')
```

```python
new_points = G.Gaussian_Generator(area,50,0.01)
plt.figure(figsize=(10,10))
V.show_single_area(area,new_points)
plt.legend(loc='lower left')
```

```python
X_generated,y_generated = cos.oversampling(X,y,min_all_safe_area,min_half_safe_area,all_safe_gen=G.Gaussian_Generator,half_safe_gen=G.Gaussian_Generator,Gaussian_scale=0.0001,minlabel=None,majlabel=None,all_safe_weight=2,show=True)
minlabel,majlabel = cos.get_labels(y)
V.show_oversampling(X, y, X_generated,y_generated)
```
