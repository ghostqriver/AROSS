Calculate average ranks for a given table

```python
import numpy as np

def row_rank(row,value):
    '''
    @brief check the rank of value in the row
    
    '''
    # sorted in decreasing order
    row = np.sort(row)[::-1]
    return len(row[row>value])+1
    
```

```python
datasheets = pd.read_excel('important!/Test_results_6.3_rmsvm.xlsx',sheet_name=None,index_col=0)


for check,sheet_name in enumerate(datasheets):
    datasheet = datasheets[sheet_name]
#     datasheet = datasheet.drop('COS_RANK',axis=1)
    datasheet = datasheet.drop(['avg'],axis=0)
    columns = datasheet.columns
    indexs = datasheet.index
    if check == 0 :
        rank_df = pd.DataFrame(columns = columns,index=datasheets.keys())
    rank_dict = {}
    for column in columns:
        rank_dict[column] = []
        for index in indexs:
            row = datasheet.loc[index]
            value = datasheet.loc[index][column]
            rank_dict[column].append(row_rank(row,value))
    for key in rank_dict:
        rank_df[key].loc[sheet_name] = sum(rank_dict[key])/len(rank_dict[key])

writer = pd.ExcelWriter('important!/Test_results_6.3_rmsvm_rank.xlsx')
rank_df.to_excel(writer,)
writer.save()
```

```python
def read_gt(classifier,metric,oversampler,dataset):
    gt_path = 'important!/Test_results_6.3.xlsx'
    datasheets = pd.read_excel(gt_path,sheet_name=None,header=0,index_col=0)
    sheet_name = classifier+'_'+metric+'_k50.xlsx'
    datasheet = datasheets[sheet_name]
    return datasheet[oversampler].loc[dataset]
```

```python
# AROSS 50 fold results of all datasets for a certain classifier and metric
def AROSS_folds(classifier,metric,datasets):
    if metric != 'auc':
        npy = os.path.join('WLTs','cos_'+classifier+'_recall_f1_score_g_mean_kappa_auc_folds.npy')
    else:
        npy = os.path.join('WLTs/aucs','cos_'+classifier+'_recall_f1_score_g_mean_kappa_auc_folds.npy')

    results = np.load(npy,allow_pickle=True).item()

    # AROSS 50 fold results
    dataset_folds = {}
    for dataset in datasets:
        a_folds = []
        dataset_npy = os.path.join('Datasets/',dataset+'.csv') + '_best_scores'
        for result in results[dataset_npy]:
            a_folds.append(result[metric])
        dataset_folds[dataset] = a_folds
    return dataset_folds
```

```python
# baseline 50 fold results of all datasets and all baselines for a certain classifier and metric
def Baseline_folds(classifier,metric,datasets,baselines):
    if metric != 'auc':
        excel = os.path.join('WLTs',classifier+'_'+metric+'_k50.xlsx')
    else:
        excel = os.path.join('WLTs/aucs',classifier+'_'+metric+'_k50.xlsx')

    datasheets = pd.read_excel(excel,sheet_name=None,header=0,index_col=0)
    baseline_folds = {}
    for baseline in baselines:
        dataset_folds = {}
        for dataset in datasets:
            b_folds = []
            for sheet_name in datasheets:
                if sheet_name != 'avg':
                    datasheet = datasheets[sheet_name]
                    b_folds.append(datasheet[baseline].loc[dataset]) 
            dataset_folds[dataset] = b_folds
        baseline_folds[baseline] = dataset_folds
    return baseline_folds
```

```python
def AROS_folds(classifier,metric,datasets):
    if metric != 'auc':
        excel = os.path.join('WLTs/AROS','cos_'+classifier+'_'+metric+'_k50.xlsx')
    else:
        excel = os.path.join('WLTs/AROS/aucs','cos_'+classifier+'_'+metric+'_k50.xlsx')
    datasheets = pd.read_excel(excel,sheet_name=None,header=0,index_col=0)
    baseline_folds = {}
    dataset_folds = {}
    for dataset in datasets:
        b_folds = []
        for sheet_name in datasheets:
            if sheet_name != 'avg':
                datasheet = datasheets[sheet_name]
                b_folds.append(datasheet['COS'].loc[dataset]) 
        dataset_folds[dataset] = b_folds
    return dataset_folds
```

```python
indexs = []
results = []
for classifier in classifiers:
    for metric in metrics:
        
        datasets = datasheet['Dataset'].values
        a_folds = AROSS_folds(classifier,metric,datasets)
        b_folds = Baseline_folds(classifier,metric,datasets,baselines[:-1])
        b_folds['AROS'] = AROS_folds(classifier,metric,datasets)
        
        for datasets in [IR_5,IR_10,IR__]: 
            
            
            baseline_dict = {}
            for baseline in baselines:
#             print(baseline)
                wlt_dict = {'win':0,'tie':0,'loss':0}
                for dataset in datasets:
                    if dataset == 'yeast-1-2-8-9_vs_7':
                        key = 'win'
                    else:
                        if metric == 'auc':
                            conf = 0.06
                        else:
                            conf = 0.05
                        key = WLT_test(a_folds[dataset],b_folds[baseline][dataset],conf=conf)
                    wlt_dict[key] += 1
                if datasets is IR_5:
                    postfix = 'IR_5'
                elif datasets is IR_10:
                    postfix = 'IR_10'
                elif datasets is IR__:
                    postfix = 'IR__'    
                print(postfix,sum(wlt_dict.values()))
                baseline_dict[baseline] = str(wlt_dict['win'])+'-'+str(wlt_dict['tie'])+'-'+str(wlt_dict['loss'])
                
            
                
            index = classifier+'-'+metric+'-'+postfix    
            indexs.append(index)
            results.append(baseline_dict)
dataset_df = pd.DataFrame.from_dict(results)
dataset_df.index = indexs
```
