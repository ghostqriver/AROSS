```python
import COS_Funcs.baseline as baseline
import COS_Funcs.cos as cos
import COS_Funcs.cure as cure
import COS_Funcs.visualize as V
from sklearn.model_selection import train_test_split
```

```python
c = 10
alpha = 0.5
N = 30


# Most different when fold = 3 on Sampledata_new_3
#            Single      |       Complete
# Recall     0.9277      |       0.9518
# All safe   24          |       30
# Half safe  11          |       17


random_state = 2    
dataset = 'Sampledata_new_3' 
```

```python
# Read the data and split
X,y = baseline.read_data('Dataset/',dataset)
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=random_state)
```

## Single

```python
cos_paras = {'c' : c,
             'alpha': alpha,
             'N': N,
             'linkage':'single'}
baseline.show_baseline(dataset,random_state,pos_label=None,args=cos_paras)
```

```python
X_generated,y_generated,_,_ = cos.COS(X_train,y_train,N,c,alpha,linkage='single',visualize=True)
```

```python
clusters,all_reps,num_reps = cure.Cure(X_train,N,c,alpha,linkage='single')
V.show_clusters_rep_points(X_train,y_train,clusters)
```
